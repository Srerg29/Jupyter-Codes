# Funciones trigonométricas

## El siguiente código grafica las funciones trigonométricas más importantes

### 
La variable <deg> indica los grados y se graficarán en su equivalente a radianes
    e.g 180° = pi radiantes

```python
import matplotlib.pyplot as plt
import numpy as np
import math

deg = 360
rad = deg*math.pi/180
x = np.arange(0,rad,rad/deg)
y = np.zeros([deg])
for i in range(len(x)):
    y[i] = math.cos(x[i])
x_deg = x*180/math.pi
plt.plot(x_deg,y)
plt.title("Plot cos(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```

```python
w = np.zeros([deg])
for i in range(len(x)):
    w[i] = math.sin(x[i])
plt.plot(x_deg,w)
plt.title("Plot sin(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show
```

```python
z = np.zeros([deg])
for i in range(len(x)):
    z[i] = math.tan(x[i])
plt.plot(x_deg,z)
plt.title("Plot tan(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show
```
