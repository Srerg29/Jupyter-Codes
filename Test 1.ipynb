```python
import matplotlib.pyplot as plt
import numpy as np
import math

deg = 20
x = np.arange(0,deg/math.pi,1/math.pi)
y = np.zeros([deg])
for i in range(len(x)):
    y[i] = math.cos(i/math.pi)
plt.plot(x,y)
plt.title("Plot cos(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```

```python
w = np.zeros([deg])
for i in range(len(x)):
    w[i] = math.sin(i/math.pi)
plt.plot(x,w)
plt.title("Plot sin(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show
```

```python
z = np.zeros([deg])
for i in range(len(x)):
    z[i] = math.tan(i/math.pi)
plt.plot(x,z)
plt.title("Plot tan(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show
```
